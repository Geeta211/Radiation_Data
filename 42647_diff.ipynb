```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn
```

```python
data = pd.read_csv('/content/Ahmedabad_diff_42647.xlsx - 1993.csv')
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
#data.set_index('timestamp', inplace=True)
```

```python
data['date'] = data['timestamp'].dt.date
data['hour'] = data['timestamp'].dt.hour
pivot_table = data.pivot_table(values='radiation', index='date', columns='hour', aggfunc='mean')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Radiation'], label='Daily Average Radiation', color='blue')
plt.title('Daily Average Radiation Over a Year')
plt.xlabel('Date')
plt.ylabel('Radiation')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
decomposition = seasonal_decompose(data['Radiation'], model='additive', period=96)
decomposition.plot()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(data['Radiation'])
plt.title('Autocorrelation of Radiation')
plt.show()
```
